# Finding and Replacing Outliers

## Skew value should be between -1 and 1 for a normal distribution

```python
print("Credit Score :",data["CreditScore"].skew())
print("Age :",data["Age"].skew())
print("Tenure :",data["Tenure"].skew())
print("Balance :",data["Balance"].skew())
print("No of Products :",data["NumOfProducts"].skew())
print("Estimated Salary :",data["EstimatedSalary"].skew())
```

```python
sns.boxplot(data["Age"])
```

### Interquartile Range IQR

```python
q1 = data["Age"].quantile(0.25)
q3 = data["Age"].quantile(0.75)
```

### Flooring and Capping

```python
iqr = q3 - q1 
whisker_width = 1.5
lower_whisker = q1 -(whisker_width*iqr)
upper_whisker = q3 + (whisker_width*iqr)
data["Age"]=np.where(data["Age"]>upper_whisker,upper_whisker,np.where(data["Age"]<lower_whisker,lower_whisker,data["Age"]))
```

```python
sns.boxplot(data["Age"])
```
